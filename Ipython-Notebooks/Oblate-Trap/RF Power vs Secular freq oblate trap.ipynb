```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
power = np.array([1.8, 2.2,2.6, 2.8])**2/4
freqs = np.array([1.133, 1.211, 1.414, 1.725])

power2 = np.array([1.4, 1.6, 1.8, 2.0, 2.2, 2.4, 2.6, 2.8])**2/4
freq2 = np.array([0.862,1.038,1.176, 1.326, 1.47, 1.567, 1.668, 1.798])

fit = np.polyfit(power2,freq2**2,1)
plt.figure()
plt.scatter(power2,freq2**2, color='r')
x = np.linspace(0,15,10)
plt.plot(x, fit[0]*x + fit[1])

plt.xlabel('Power (Watts)')
plt.ylabel('Frequency^2 (no 2 pi)')
plt.title('Secular Frequencies^2 vs. Power')
plt.savefig("/home/qsimexpcontrol/plots/oblatemodes/freqvspowerwithfit.pdf")
plt.show()
```
