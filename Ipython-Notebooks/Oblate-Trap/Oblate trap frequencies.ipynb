```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

xymodes1 = np.loadtxt('/home/qsimexpcontrol/plots/oblatemodes/tickle_scans/669.csv',delimiter=',')
xymodes2 = np.loadtxt('/home/qsimexpcontrol/plots/oblatemodes/tickle_scans/670.csv',delimiter=',')
zmode = np.loadtxt('/home/qsimexpcontrol/plots/oblatemodes/tickle_scans/675.csv',delimiter=',')
```
