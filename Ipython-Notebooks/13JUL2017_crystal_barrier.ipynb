```python
def do_drags(assy):
    set_assymetry(assy)
    E_copy, fcp_copy=make_crystal(starting_ions=fcp)
    fcp_copy[0].color='red'
    fcp_copy[2].color='green'
    fcp_copy[5].color='green'
    plot_ions(fcp_copy) 
    x_mid_point = (fcp_copy[2].x + fcp_copy[5].x)/2
    y_mid_point = (fcp_copy[2].y + fcp_copy[5].y)/2
    mid = [(x_mid_point, y_mid_point)]
    coords, E = drag_ion(crystal = fcp_copy, initpos=[(fcp_copy[0].x, fcp_copy[0].y)], finalpos=mid, steps = 200)
    return coords, E

squeezes = np.linspace(1,1.2, 10)


E_list = []

for squeeze in squeezes:
    coord, E = do_drags(squeeze)
    E_list.append(E)
```
