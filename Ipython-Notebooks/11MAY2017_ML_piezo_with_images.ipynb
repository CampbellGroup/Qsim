```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def data_average(preamble, experiment, datasets):
    for i, item in enumerate(datasets):
        item = str(item)
        if len(item) < 5:
            item = (5 -len(item))*'0' + item
        temp = np.loadtxt(preamble + item  + ' - ' + experiment + '.csv', delimiter=',')
        if i != 0:
            prev_y = temp[:,1] + prev_y
        else:
            x_values = temp[:,0]
            prev_y = temp[:,1]
    y_values = prev_y/len(datasets)
    return [x_values, y_values]
```

## Taken with 20 pics CW on 92 $\mu W$

```python
x, y = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range(438,447))
images = np.load('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/00438 - MLpiezoscan.npy')
image_list = []
for item in images:
    print np.size(item)
    data_resize = item.reshape(300, 300)
    image_list.append(data_resize)
np.load()
plt.plot(x, y)
for image in image_list:
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.imshow(image, clim=[300, 400], cmap='gray')
```
