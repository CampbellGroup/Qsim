```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
```

```python
from qutip import *
```

```python
# Hamiltonian
rabi = 1.0
detuning = 0.0
H = sigmax()*np.sqrt(rabi**2)/2.0 + sigmaz()*detuning

# initial state
psi0 = basis(2, 1)

# list of times for which the solver should store the state vector
tlist = np.linspace(0, np.pi*rabi, 100)

meas = (sigmaz() + qeye(2))/2.0
result = mesolve(H, psi0, tlist, [], [meas])
```

```python
fig, axes = plt.subplots(1,1)

axes.plot(tlist, result.expect[0])

axes.set_xlabel(r'$t$', fontsize=20)
axes.set_ylabel(r'$\left<\sigma_z\right>$', fontsize=20);
```

```python
freq_list = np.linspace(-10.*rabi, 10.0*rabi, 80)
tlist = np.linspace(0, 3.5*np.pi*rabi, 2)
amplitudes=[]
psi = basis(2,1)
for detuning in freq_list:
    H = sigmax()*np.sqrt(rabi**2)/2.0 + sigmaz()*detuning
    result = mesolve(H, psi, tlist, [], [meas])
    amplitudes.append(result.expect[0][-1])
plt.plot(freq_list, amplitudes)
```
