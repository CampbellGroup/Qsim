```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import time

amurange = 100
path = '/share/Qsim/RGA/2nd-init-ti-sub-Bake'
files = os.listdir(path)
data = np.zeros((991,len(files)))
times = []
plt.xlim(1,40)
for i, file in enumerate(files):
    with open(path+file, 'r') as f:
        timestamp = f.readline()
    timestamp = time.strptime(timestamp[0:25], "%b %d, %Y %I:%M:%S %p")
    times.append(time.mktime(timestamp))
    tmp = np.loadtxt(path + file, skiprows= 22, dtype = 'string')
    tmp = np.char.strip(tmp,',')
    data[:,i] = tmp[:,1]
    plt.semilogy(tmp[:,0], data)
plt.savefig("/share/Qsim/RGA/2nd-init-ti-sub-Bake/first-post-tisubBakeout-spectra.pdf")
```

```python
amutoplot = [2,16,28,32,40]
width = 2
res = 10
for amu in amutoplot:
    minpoint = res*(amu - 1 - width/2)
    if minpoint < 0: minpoint = 0
    maxpoint = res*(amu - 1 + width/2)
    if maxpoint > amurange*res: maxpoint = amurange
    amudata = data[minpoint:maxpoint,:] 
    amudata = sum(amudata,0)/len(amudata)
    x = np.array(times)
    y = amudata
    indexs_to_order_by = x.argsort()
    x_ordered = x[indexs_to_order_by]
    x_ordered = (x_ordered - x_ordered[0])/3600
    y_ordered = y[indexs_to_order_by]
    plt.semilogy(x_ordered,y_ordered)
plt.xlabel('Time (hrs)')
plt.ylabel('Pressuer (mbar)')
plt.title('First post ti-sub bakeout @ 200 C')
#plt.savefig("/share/Qsim/RGA/2nd-init-ti-sub-Bake/first-post-tisubBakeout.pdf")
plt.legend(amutoplot,3)
```
