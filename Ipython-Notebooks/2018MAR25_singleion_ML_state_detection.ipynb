```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('crizzle')
import csv

path = '/home/qsim-rad/LabRAD/data/'

#uWlinescanbad = '/home/qsim-rad/Documents/uWlinescan_ml_statedetection(outofphase).csv'
#rabiFlopping = '/home/qsim-rad/Documents/rabi_flopping_ml_statedetection(outofphase).csv'
#uWcsv = open(rabiFlopping, 'w')
```

```python
uWscans = [3613, 3614, 3615, 3619, 3620, 3621]
rabiFlops = [206, 207]

uW = []
rabi = []
for num in uWscans:
    ss = path +'/MicrowaveLineScan.dir/0%s - MicrowaveLineScan.csv' % (str(num))
    uW.append(np.loadtxt(ss, delimiter = ','))
    
for num in rabiFlops:
    ss = path +'/MicrowaveRabiFlopping.dir/00%s - MicrowaveRabiFlopping.csv' % (str(num))
    rabi.append(np.loadtxt(ss, delimiter = ','))

uWavg = np.zeros(50)
for i in range(5):
    uWavg += uW[i][:,1]
uWavg = uWavg/5.

#for n in range(100):
#    row = str(rabi[1][:,0][n]) +',' + str(rabi[1][:,1][n])
#    uWcsv.write(row +'\n')
#print len(rabi[1][:,0])
```

```python
plt.plot(uW[1][:,0],uWavg , label='Interferometer in phase')
plt.plot(uW[5][:,0], uW[5][:,1], label='Interferometer out of phase', color='g')
plt.legend(loc=1)
plt.ylabel(r'$P_{ex}')
plt.xlabel(r'Microwave detuning from 12.642812 GHz [kHz]')
#plt.savefig('/home/qsim-rad/Documents/single_ion_uW_linescan_ML.pdf')
```
