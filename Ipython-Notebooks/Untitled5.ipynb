```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 1000)
sin_1 = np.sin(x)
sin_2 = np.sin(x + np.pi/2)
plt.plot(x, sin_1**2)
plt.plot(x, sin_2**2)
```

```python
sin_3 = sin_1**2 + sin_2**2
plt.plot(x, sin_3)
```
