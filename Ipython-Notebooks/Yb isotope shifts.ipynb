```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
c = 2.998e8
def lorentzfunc(x, x0, gamma, A):
   return (A/(2*np.pi)) * gamma/((x - x0)**2 + (gamma/2)**2)
```

```python
yb_369_wl = {170: 369.5236, 171: 369.5261, 172: 369.5246, 174: 369.5250, 176: 369.5255}
yb_935_wl = {170: 935.1982, 171: 935.1878, 172: 935.1874, 174: 935.1799, 176: 935.1724}
yb_399_wl = {176: 398.91178, 174: 398.9118, 172: 398.9116, 171: 398.9113, 170: 398.9111}
```

```python
yb_369_freq = {}
yb_935_freq = {}
yb_399_freq = {}
for isotope in yb_369_wl.keys():
    yb_369_freq[isotope] = c/yb_369_wl[isotope]
    
for isotope in yb_935_wl.keys():
    yb_935_freq[isotope] = c/yb_935_wl[isotope]
    
for isotope in yb_369_wl.keys():
    yb_399_freq[isotope] = c/yb_399_wl[isotope]
    
yb_369_freq[173] = yb_369_freq[174] -.740
yb_399_freq[173] = yb_399_freq[174] + .590
yb_935_freq[173] = yb_935_freq[174] -.500
```

```python
x = np.linspace(-0.5, 1.75, 20000)
for isotope in yb_399_freq.keys():
    center = yb_399_freq[isotope] - yb_399_freq[174]
    y = lorentzfunc(x, center,0.02, 1.0)
    plt.plot(x,y, label=isotope)
plt.legend()
```

```python
x = np.linspace(-5.6, 5.3, 20000)
for isotope in yb_369_freq.keys():
    center = yb_369_freq[isotope] - yb_369_freq[174]
    y = lorentzfunc(x, center,0.02, 1.0)
    if isotope == 171:
        y = y + lorentzfunc(x, center + 7.374, 0.02, 0.75) + lorentzfunc(x, center - 7.374, 0.02, 0.75) + lorentzfunc(x, center - 2.105, 0.02, 0.75) + lorentzfunc(x, center + 2.105, 0.02, 0.75)
    plt.plot(x,y, label=isotope)
plt.title('369 Isotope shifts')
plt.legend(loc=3)
```

```python
%matplotlib notebook
x = np.linspace(-7.5, 4, 20000)
for isotope in yb_935_freq.keys():
    center = yb_935_freq[isotope] - yb_935_freq[174]
    y = lorentzfunc(x, center,0.02, 1.0)
    if isotope == 171:
        y = y + lorentzfunc(x, center + 3.06, 0.02, 0.75) + lorentzfunc(x, center - 3.06, 0.02, 0.75)
    plt.plot(x,y, label=isotope)
plt.legend()
```
