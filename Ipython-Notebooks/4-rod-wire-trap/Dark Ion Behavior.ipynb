```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fiveion = np.array([['4:47:21', 2, 28], ['4:47:42', 1, 13], ['4:50:44', 2, 75], ['4:51:47', 1, 23], ['4:51:51', 2, 1],
                    ['4:52:25', 1, 18], ['4:52:35', 2, 2], ['4:55:15', 3, 69], ['4:56:15', 4, 23], ['4:57:36', 3, 19], 
                    ['4:57:43', 2, 9], ['4:58:05', 1, 6], ['4:58:26', 2, 9], ['4:59:51', 3, 33], ['5:00:24', 4, 17], 
                    ['5:01:08', 3, 10], ['5:01:20', 4, 4], ['5:01:30', 3, 3], ['5:02:00', 2, 10], ['5:03:23', 3, 30], 
                    ['5:03:46', 4, 3], ['5:03:57', 3, 5], ['5:05:04', 4, 20], ['5:05:55', 3, 15], ['5:06:41', 2, 15],
                    ['5:08:21', 1, 42], ['5:08:39', 0, 5], ['5:09:33', 1, 0], ['5:10:03', 0, 4], ['5:10:03', 7, 0]])

fourion = np.array([['5:34:12',3, 0], ['5:35:12', 2, 10], ['5:35:13',1, 1], ['5:35:14', 2, 0], ['5:35:15', 1, 0], 
                   ['5:35:17', 2, 0], ['5:36:21', 3, 12], ['5:36:31', 2, 2], ['5:37:44', 1, 13], ['5:38:07', 2, 4],
                   ['5:38:16', 1, 1], ['5:38:23',0, 0], ['5:38:23', 7, 0]])


def convertarray(arr):
    start = getSec(arr[0,0])
    for i, v in enumerate(arr[:,0]):
        arr[i,0] = getSec(v) - start
        
def getSec(s):
    l = s.split(':')
    return int(l[0]) * 3600 + int(l[1]) * 60 + int(l[2])
convertarray(fiveion)
convertarray(fourion)

fiveion = fiveion.astype(np.int)
fourion = fourion.astype(np.int)

totaltimefive = int(fiveion[-1,0]) -int(fiveion[0,0])
totaljumpsfive = np.sum(fiveion[:,2], dtype = np.int32)
jumpsperminfive = 60.*totaljumpsfive/totaltimefive
totaltimefour = int(fourion[-1,0]) -int(fourion[0,0])
totaljumpsfour = np.sum(fourion[:,2], dtype = np.int32)t
jumpsperminfour = 60.*totaljumpsfour/totaltimefour

print "jumps per min five ions = " + str(jumpsperminfive)
print "average seconds to change state five ions = " + str(float(totaltimefive)/(np.size(fiveion[:,0])))
print "jumps per min four ions = " + str(jumpsperminfour)
print "average seconds to change state four ions = " + str(float(totaltimefour)/(np.size(fourion[:,0])))

hopsfive = []
timeinstatefive = []
hopsfour = []
timeinstatefour = []

for i in range(5):
        timeforionfive = fiveion[np.where(fiveion[:,1] == i)[0]+1][:,0]
        timeforprevionfive = fiveion[np.where(fiveion[:,1] == i)[0]][:,0]       
        hopsforionfive = fiveion[np.where(fiveion[:,1] == i)[0]+1][:,2]
        timeinconfigfive =  timeforionfive - timeforprevionfive
        hopsfive.append(sum(hopsforionfive)/len(hopsforionfive))
        timeinstatefive.append(sum(timeinconfigfive)/len(hopsforionfive))
        
for i in range(4):
        
        timeforionfour = fourion[np.where(fourion[:,1] == i)[0]+1][:,0]
        timeforprevionfour = fourion[np.where(fourion[:,1] == i)[0]][:,0]       
        hopsforionfour = fourion[np.where(fourion[:,1] == i)[0]+1][:,2]
        timeinconfigfour =  timeforionfour - timeforprevionfour
        hopsfour.append(sum(hopsforionfour)/len(hopsforionfour))
        timeinstatefour.append(sum(timeinconfigfour)/len(hopsforionfour))

fig, ax = plt.subplots()       
bar1 = ax.bar(np.arange(len(hopsfive)) - 1, hopsfive, width=.5, color = 'b')
bar2 = ax.bar(np.arange(len(hopsfour)) - 0.5, hopsfour, width=.5, color = 'g')

ax.set_title("Ion chain hopping")
ax.set_xticks(np.arange(5) + 0.5)
ax.set_xticklabels( ('1', '2', '3', '4', '5') )
ax.legend(('Five Ions', 'Four Ions'))
plt.xlabel("Number of (N) bright ions")
plt.ylabel("Number of hops per minute with N ions")
plt.show()

fig, ax = plt.subplots() 
bar1 = ax.bar(np.arange(len(timeinstatefive)) - 1, timeinstatefive, width=.5, color = 'b')
bar2 = ax.bar(np.arange(len(timeinstatefour)) - 0.5, timeinstatefour, width=.5, color = 'g')
ax.set_title("Time in configuration")t
ax.set_xticks(np.arange(5) + 0.5)
ax.set_xticklabels( ('1', '2', '3', '4', '5') )
ax.legend(('Five Ions', 'Four Ions'))

plt.xlabel("Number of (N) bright ions")
plt.ylabel("Average time with N ions (seconds)")
plt.show()

fig, ax = plt.subplots()
plot1 = plt.plot(fiveion[0:-1,0],fiveion[0:-1,1], fourion[0:-1,0],fourion[0:-1,1], lw = 2)
plt.xlabel("Elapsed Time (s)")
plt.ylabel("Number of Ions")
ax.legend(('Five Ions', 'Four Ions'))
plt.show()





  
```
