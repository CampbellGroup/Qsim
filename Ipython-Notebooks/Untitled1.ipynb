```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
   def data_average(preamble, experiment, datasets):
    for i, item in enumerate(datasets):
        item = str(item)
        if len(item) < 5:
            item = (5 -len(item))*'0' + item
        temp = np.loadtxt(preamble + item  + ' - ' + experiment + '.csv', delimiter=',')
        if i != 0:
            prev_y = temp[:,1] + prev_y
        else:
            x_values = temp[:,0]
            prev_y = temp[:,1]
    y_values = prev_y/len(datasets)
    return [x_values, y_values]
```

```python
x1, y1 = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range(639,644))
x2, y2 = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range(678,688))
```
