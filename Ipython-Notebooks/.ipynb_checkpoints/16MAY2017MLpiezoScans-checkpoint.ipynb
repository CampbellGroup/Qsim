```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
pylab.rcParams['figure.figsize'] = (12,8)
```

```python
def data_average(preamble, experiment, datasets):
    for i, item in enumerate(datasets):
        item = str(item)
        if len(item) < 5:
            item = (5 -len(item))*'0' + item
        temp = np.loadtxt(preamble + item  + ' - ' + experiment + '.csv', delimiter=',')
        if i != 0:
            prev_y = temp[:,1] + prev_y
        else:
            x_values = temp[:,0]
            prev_y = temp[:,1]
    y_values = prev_y/len(datasets)
    return [x_values, y_values]
```

```python
x1, y1 = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range(825,827))
x2, y2 = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range(831,833))
x3, y3 = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range(834,836))
x4, y4 = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range60(839,841))
x5, y5 = data_average('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/', 'MLpiezoscan', range(842,844))
```

```python
plt.plot(x1[1:],y1[1:], label = '1')
plt.plot(x2[1:] + 0.5,y2[1:], label = '2')
plt.plot(x3[1:] + 0.53,y3[1:], label = '3')
plt.plot(x4[1:] + 1,y4[1:], label = '4')
plt.plot(x5[1:] + 1.1,y5[1:], label    = '5')
plt.legend(loc=0)
```

```python
x =[]
y = []
for item in [404, 406, 407, 408, 409]:
    x_temp, y_temp = data_average('/home/qsimexpcontrol/LabRAD/data/scan_935.dir/', 'scan_935', range(item,item+1))
    x = x + list(x_temp)
    y = y + list(y_temp)
s = sorted(zip(x,y))
x,y = map(list, zip(*s))
points = 90
x_avg = np.linspace(min(x),max(x), points)
y_avg = []
for i, point in enumerate(x_avg):
    if i != points - 1:
        window = np.where(np.logical_and(x>=point, x<x_avg[i+1]))[0]
        temp = [y[i] for i in window]
        y_temp = np.sum(temp)/len(temp)
        y_avg.append(y_temp)

#plt.plot(x,y)
plt.plot(x_avg[15:-1]*2, np.max(y_avg[15:]) - y_avg[15:], linewidth= 5.0, color='forestgreen')
plt.title('2S1/2 to 2D5/2 Transition in 174 Yb+ at 411nm', fontsize=25)
plt.ylabel('Fluorescence dip (kcounts/s)', fontsize= 20)
plt.xlabel('Detuning from 729.475120 THz (MHz)', fontsize=20)
```
