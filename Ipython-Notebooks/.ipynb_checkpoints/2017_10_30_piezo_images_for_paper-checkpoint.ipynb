```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.ndimage.interpolation import rotate
import matplotlib.gridspec as gridspec
```

```python
images = np.load('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/01161 - MLpiezoscan.npy')
image_list = []
for item in images:
    data_resize = item.reshape(401, 401)
    image_list.append(data_resize)
```

```python
%matplotlib inline
res = 2.5/16
fig = plt.figure(figsize=(120.5,50))
gs1 = gridspec.GridSpec(4, 4)
gs1.update(wspace=0.0, hspace=0.2)
for i, image in enumerate(image_list):
    volt = 2.7 + i*res
    ax = plt.subplot(gs1[i])
    rotated = rotate(image, 65, reshape=False)
    cropped = rotated[150:250,80:321] 
    ax.imshow(cropped, cmap='gray', clim=[600,850])
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    ax.set_adjustable('box-forced')
    ax.annotate(str(volt) + 'V', (0,0), (0, -20), xycoords='axes fraction', textcoords='offset points', va='top', size=65)
plt.savefig('/home/qsimexpcontrol/plots/phonon_laser/2017_10_30_scan_images.pdf')
```
