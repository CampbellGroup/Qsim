```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
mlPiezo137 = np.loadtxt('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/00137 - MLpiezoscan.csv', delimiter=",")
mlPiezo138 = np.loadtxt('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/00138 - MLpiezoscan.csv', delimiter=",")
mlPiezo147 = np.loadtxt('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/00147 - MLpiezoscan.csv', delimiter=",")
mlPiezo148 = np.loadtxt('/home/qsimexpcontrol/LabRAD/data/MLpiezoscan.dir/00148 - MLpiezoscan.csv', delimiter=",")
```

```python
volts40 = mlPiezo137[:,0][1:200]
volts30 = mlPiezo137[:,0][1:200]
val137 = mlPiezo137[:,1][1:200]
val138 = mlPiezo138[:,1][1:200]
val147 = mlPiezo147[:,1][1:200]
val148 = mlPiezo148[:,1][1:200]
plt.ylabel('Intensity (kCounts/sec)')
plt.xlabel('Tsunamo Piezo Voltage (V)')
valAvg3 = []
valAvg4 = []
for counts in range(len(val137)):
   valAvg3.append((val137[counts] + val138[counts])/2.0)
   valAvg4.append((val147[counts] + val148[counts])/2.0) 
plt.title('ML line scan NO CW')
plt.plot(volts30,valAvg3, label='200 microW ML')
plt.plot(volts40,valAvg4, label='400 microW ML')
plt.legend(loc=3)
#plt.savefig('/home/qsimexpcontrol/plots/ML_Laser/8MAY2017ML_only_linescan.png')
```
