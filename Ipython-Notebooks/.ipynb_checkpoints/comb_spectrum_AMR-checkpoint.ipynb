```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
order = 10
beta_0 = 1.2
deltas = np.linspace(-100, 100, 1000)
gamma = 20
trap_freq = 48.5


def spectrum(deltas, gamma, trap_freq, beta_0, order, offset):
    P = np.zeros(len(deltas))
    y_values = []
    for detuning in deltas:
        tmp = 0
        for n in range(-1*order, order):
            tmp += scipy.special.jn(n,beta_0)**2 / ((detuning - offset + n*trap_freq)**2 + (gamma/2 + 0.1*gamma*beta_0)**2)
        y_values.append(tmp)
    return y_values
Y_array = []
for i in range(-5,5):
    y_values = spectrum(deltas, gamma, trap_freq, beta_0, order, i*80)
    Y_array.append(y_values)
    plt.plot(deltas, y_values, alpha = 0.3)
tot_y = np.zeros(1000)
for array in Y_array:
    tot_y += array
plt.plot(deltas, tot_y)
```

```python
con = np.convolve(Y_array[0], Y_array[1], 'same')
plt.plot(deltas, con)
```
