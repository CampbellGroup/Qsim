```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
order = 10
beta_0 = 1.1
delta = np.linspace(-300, 300, 1000)
gamma = 35.0
trap_freq = 48.5
P = np.zeros(1000)
y_values = []
for detuning in delta:
    tmp = 0
    if detuning <=0:
        beta = 0.0
    elif beta <=:
        beta = beta_0*detuning/20
    for n in range(-1*order, order):
        tmp += scipy.special.jn(n,beta)**2 / ((detuning + n*trap_freq)**2 + (gamma/2)**2)
    y_values.append(tmp)
plt.plot(delta, y_values)
```
