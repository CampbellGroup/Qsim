```python
pmtdata = np.loadtxt('/home/qsimexpcontrol/LabRAD/data/PMT Counts.dir/01150 - PMT Counts.csv',delimiter=',')
```

```python
def coursen(xdata,ydata,points):
    newlength = int(len(xdata)/points)
    newxdata = []
    newydata = []
    j = 0
    for i in range(newlength):
        i = points*(i)
        newydata.append(np.mean(ydata[int(j):int(i)]))
        newxdata.append(xdata[int((i + j)/2)])
        j = i
    return newxdata, newydata
```

```python
time = pmtdata[12150:30500][:,0]
counts = pmtdata[12150:30500][:,3]
coursedx, coursedy = coursen(np.array(time - time[0]),np.array(counts), 200)
plt.title('Ion lifetime ML laser only')
plt.xlabel('Time (s)')
plt.ylabel('Intensity (kCounts/second)')
plt.plot(coursedx, coursedy, color='g', linewidth=2)
#plt.savefig('/home/qsimexpcontrol/plots/ML_Laser/8MAY2017ML_only_lifetime.png')
```
