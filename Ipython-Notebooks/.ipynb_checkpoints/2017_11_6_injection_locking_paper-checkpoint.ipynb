```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator

mpl.rcParams['axes.linewidth'] = 3.


path = '/home/qsimexpcontrol/LabRAD/data/PMT_FFT.dir/'

%matplotlib inline
```

```python
timeRes = 10*10**-9
timeObs = 0.4

data301 = np.loadtxt(path + '01147 - PMT_FFT.csv', delimiter = ',')
data302 = np.loadtxt(path + '01149 - PMT_FFT.csv', delimiter = ',')

freq301, y301 = data301[:,0]/1000., data301[:,1]*timeObs/50.
freq302, y302 = data302[:,0]/1000., data302[:,1]*timeObs/50.
```

```python
fig = plt.figure(figsize = (12,8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize = 15)


plt.plot(freq301[:700], 10*np.log10(1000*y301[:700]) + 58., 'purple')
plt.plot(freq302[:700], 10*np.log10(1000*y302[:700]) + 58., 'g')

plt.xlabel('Frequency (kHz)', fontsize = 22)
plt.ylabel(r'Power (dB/Hz)', fontsize = 22)
plt.savefig(path + 'Injection_Locking_Plot.pdf', bbox_inches="tight")
plt.xlim([543, 552])
```
