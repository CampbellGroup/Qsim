```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def coursen(xdata,ydata,points):
    newlength = int(len(xdata)/points)
    newxdata = []
    newydata = []
    j = 0
    for i in range(newlength):
        i = points*(i)
        newydata.append(np.mean(ydata[int(j):int(i)]))
        newxdata.append(xdata[int((i + j)/2)])
        j = i
    return newxdata, newydata
```

```python
#data = np.loadtxt('/home/qsimexpcontrol/phase_paper/95_meter_fiber_10_MHz_sweep.csv',delimiter=',')
data = np.genfromtxt('/home/qsimexpcontrol/phase_paper/95_meter_fiber_10_MHz_sweep.csv', delimiter=',')[:,:-1]
x = data[0:,0]
y = data[0:, 1]
xwindow = x[5030:8690]
ywindow = y[5030:8690]
xcoarse, ycoarse = coursen(xwindow, ywindow, 15)
scaled_y = np.array(ycoarse) - 0.0015*np.array(xcoarse)
scaled_y = scaled_y/np.max(scaled_y[1:])
scaled_x = 310.0 + 5/2. *np.array(xcoarse)
#plt.xlim([310.0, 320.0])
#plt.ylim([0.80, 1.02])
plt.title('95 Meter Interferometer')
plt.xlabel('Frequency (MHz)')
plt.ylabel('Intensity (A.U.)')
plt.plot(scaled_x,scaled_y, 'g', linewidth = '3')
plt.savefig('/home/qsimexpcontrol/phase_paper/95_meter_plot.png')
```
