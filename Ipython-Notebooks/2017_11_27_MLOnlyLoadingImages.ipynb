```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/home/qsimexpcontrol/LabRAD/data/quick_images.dir/'
image_nums = [59, 53, 54, 60, 61, 62]
image = np.load(path + '00059 - quick_images.npy')
data_resize = image.reshape(1002, 1004)
crop_image = data_resize[250:500,600:900]
image_list = []
for num in image_nums:
    im = np.load(path + '000' + str(num) + ' - quick_images.npy')
    im_resize = im.reshape(1002, 1004)
    crop_im = im_resize[250:500,600:900]
    image_list.append(crop_im)
```

```python
plt.imshow(crop_image, clim = [800, 900])
```

```python
fig = plt.figure(figsize=(50, 50))
for i, image in enumerate(image_list):
    ax = plt.subplot2grid((1,7),(0, i))
    ax.set_axis_off()
    ax.imshow(image, clim=[700, 1000])
plt.subplots_adjust(wspace=0, hspace=0)
plt.savefig(path + 'ML_only_loaded_in_sequence.pdf')
```
