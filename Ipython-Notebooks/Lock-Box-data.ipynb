```python
def coursen(xdata,ydata,points):
    newlength = int(len(xdata)/points)
    newxdata = []
    newydata = []
    j = 0
    for i in range(newlength):
        i = points*(i)
        newydata.append(np.mean(ydata[int(j):int(i)]))
        newxdata.append(xdata[int((i + j)/2)])
        j = i
    return newxdata, newydata
```

```python
#plt.ylim([0,.00025])
plt.xlim([0,350])
plt.ylim([1e-12,1e-6])
plt.title('Locked FFT of Interferometer with new diode')
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'Power Spectral density $(V^2/Hz)$')
plt.semilogy(xf, T*2.0/N * np.abs(unlocked3fft[0:N/2])**2 , label = 'unlocked data set 3')
plt.semilogy(xf, T*2.0/N * np.abs(locked3fft[0:N/2])**2 , label = 'locked data set 3')
#plt.plot(xf, 2.0/N * np.abs(lockedlowgainfft[0:N/2]**2), label = 'Locked with gain Down')
#plt.legend(loc=9)
#plt.savefig('/home/tony/lockboxdata/seconddiode/FFTplotnewdiode.pdf')
```

```python
#plt.ylim([0,.00025])
plt.xlim([0,340])
plt.ylim([1e2,1e4])
plt.title('Locked FFT of Interferometer with new diode')
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'Power Spectral density $ \frac{kHz}{\sqrt{Hz}} $')
xfyunlocked = 2.2e7*np.sqrt(T * 2.0/N * np.abs(unlocked3fft[0:N/2])**2)
xfylocked = 2.2e7*np.sqrt(T * 2.0/N * np.abs(locked3fft[0:N/2])**2)

newxunlocked, newyunlocked = coursen(xf,xfyunlocked,10)
newxlocked, newylocked = coursen(xf,xfylocked,10)
plt.semilogy(newxunlocked, np.array(newyunlocked) , label = 'unlocked data set 3')
plt.semilogy(newxlocked,np.array(newylocked) , label = 'locked data set 3')
plt.savefig('/home/qsimexpcontrol/plots/Lock_Box/scaledlockcomparison.pdf')
#plt.plot(xf, 2.0/N * np.abs(lockedlowgainfft[0:N/2]**2), label = 'Locked with gain Down')
#plt.legend(loc=9)
```
