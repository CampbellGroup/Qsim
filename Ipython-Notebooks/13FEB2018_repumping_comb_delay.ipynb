```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def data_average(preamble, experiment, datasets):
    for i, item in enumerate(datasets):
        item = str(item)
        if len(item) < 5:
            item = (5 -len(item))*'0' + item
        temp = np.loadtxt(preamble + item  + ' - ' + experiment + '.csv', delimiter=',')
        if i != 0:
            prev_y = temp[:,1] + prev_y
        else:
            x_values = temp[:,0]
            prev_y = temp[:,1]
    y_values = prev_y/len(datasets)
    return [x_values, y_values]
```

```python
sets = [179, 180, 181, 182]
x, y = data_average('/home/qsimexpcontrol/LabRAD/data/Delaystagescan.dir/', 'Delaystagescan', sets)
plt.plot(x,y, 'orange')
plt.xlabel('piezo voltage (Volts)')
plt.ylabel('fluorescence (kcts/s)')
plt.title('delay stage scan 3 ions with weak repumping comb')
plt.savefig('/home/qsimexpcontrol/plots/ML_Laser/13FEB2018Ramsey_repump_ion.pdf')
```
