```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
center_pll_dds = 12.6428039223e9
center_pll_magiq = 12.64260186e9
DDS_source_wide = np.loadtxt('/home/qsimexpcontrol/plots/DDS_Noise/Trace_0011_data.csv',delimiter=',')
DDS_source_narrow = np.loadtxt('/home/qsimexpcontrol/plots/DDS_Noise/Trace_0012_data.csv',delimiter=',')
magiQ_source_narrow = np.loadtxt('/home/qsimexpcontrol/plots/DDS_Noise/Trace_0018_data.csv',delimiter=',')
magiQ_source_wide = np.loadtxt('/home/qsimexpcontrol/plots/DDS_Noise/Trace_0017_data.csv',delimiter=',')
```

```python
plt.figure()
plt.subplot(121)
plt.plot((DDS_source_wide[:,0]-center_pll_dds)/1000., DDS_source_wide[:,1], 'purple', label='source=DDS')
plt.plot((magiQ_source_wide[:,0]-center_pll_magiq)/1000., magiQ_source_wide[:,1], 'darkorange', alpha=0.8, label='source=MagiQ')
plt.xlabel('Detuning from 12.642 GHz (kHz)')
plt.ylabel('Power (dBm)')
plt.legend()

plt.subplot(122)
plt.plot((DDS_source_narrow[:,0] - center_pll_dds)/1000., DDS_source_narrow[:,1], 'purple', label='source=DDS')
plt.plot((magiQ_source_narrow[:,0]-center_pll_magiq)/1000., magiQ_source_narrow[:,1], 'darkorange', alpha=0.8, label='source=MagiQ')
plt.xlabel('Detuning from 12.642 GHz (kHz)')
plt.ylabel('Power (dBm)')
plt.suptitle('PLL output DDS vs MagiQ', fontsize=16)
plt.legend()
#plt.savefig("/home/qsimexpcontrol/plots/DAC_Noise/before_afterextfilter.pdf")
plt.show()
```
